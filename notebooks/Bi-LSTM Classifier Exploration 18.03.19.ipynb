```python
import sys 
sys.path.append("..")
from src.preprocess.data_prep_offenseval import DataPrepOffensEval
from src.classifiers.classifier_bi_lstm import BiLstmClassifier
from src.feature_extraction.w2i import w2i
from src.utils.stats import get_distribution_from_y
import pandas as pd
```

```python
# Get the training data
english_train_file_path = "../data/raw/OffensEval2019/start-kit/training-v1/offenseval-training-v1.tsv"
dp = DataPrepOffensEval()
result_tuple = dp.get_X_and_ys(file_path=english_train_file_path)
X_original = result_tuple[0]
y_sub_a = result_tuple[1]
```

```python
print(X_original)
```

```python
print(y_sub_a)
```

```python
# Take the whole corpus and create a word-2-index dictionary.
# Create vectors, of length equal to the longest tweet, for each tweet, where each number represents the word index \
# from the newly created dictionary
# The 'w2i' function tokenizes each tweet and removes stopwords (using nltk's stopwords module)
X, w2i_dict, i2w_dict = w2i(X_original)
print(X_original[0])
print("X[0]: {}".format(X[0]))
print("len(X): {}, len(X[0]): {}".format(len(X), len(X[0])))
print(w2i_dict.keys())
print(i2w_dict.keys())
```

```python
# Turn X and y into pandas dataframes to make it easier to keep track of indices
X = pd.DataFrame(X)
print(X.shape)
y = pd.DataFrame(y_sub_a)
print(y.shape)
```

```python
# Create train, validation and test sets:
X_train, X_test, y_train, y_test = dp.train_test_split(X, y, test_size=0.2)
X_train, X_val, y_train, y_val = dp.train_test_split(X_train, y_train, test_size=0.1)
```

```python
# Get distribution of 0/1 in different y's, to make sure all of them have similar amounts of OFF and NOT
print("y_train: {}".format(get_distribution_from_y(y_train)))
print("y_val: {}".format(get_distribution_from_y(y_val)))
print("y_test: {}".format(get_distribution_from_y(y_test)))
```

```python
# Model
from tensorflow import keras
from tensorflow.keras import initializers
from tensorflow.keras import regularizers
from tensorflow.keras.optimizers import SGD
```

```python
import tensorflow as tf
# model = None
model = keras.Sequential()
# 1) Embedding layer that learns embeddings for each word as we train. 
# The shape of the matrix is len(X[0]) x (desired_emb_dimension) == input_dim x output_dim == (rowsXcols)
model.add(keras.layers.Embedding(
    input_dim=len(w2i_dict),
    output_dim=300,
#     embeddings_initializer=initializers.glorot_normal(seed=None),
#     embeddings_regularizer=regularizers.l2(0.01),
#     activity_regularizer=regularizers.l1(0.01),
))

# 2) Dropout 
# model.add(keras.layers.Dropout(0.5))

# 3) Bi-LSTM 
model.add(keras.layers.Bidirectional(
    layer=keras.layers.LSTM(
        units=50, # Dimensionality of the output space
#         kernel_regularizer=regularizers.l1(0.01),
#         recurrent_regularizer=regularizers.l2(0.01),
#         dropout=0.1,
#         recurrent_dropout=0.1,
#         return_sequences=False,
#         unit_forget_bias=True,
        
    ),
    merge_mode="concat") # Concat the output from fwd & bwd lstm's 
)

# 4) Dropout
# model.add(keras.layers.Dropout(0.1))

# 5) Fully connected layer
model.add(keras.layers.Dense(
    units=64, 
    activation=tf.nn.relu,
#     kernel_initializer=initializers.glorot_normal,
#     kernel_regularizer=regularizers.l2(0.01),
#     activity_regularizer=regularizers.l1(0.01),
))

# 6) Dropout
model.add(keras.layers.Dropout(0.1))

# 7) Fully connected layer
model.add(keras.layers.Dense(
    units=1, 
    activation="sigmoid",
))


# -- Compile --
model.compile(
    optimizer=SGD(lr=0.5),
    loss="binary_crossentropy",
    metrics=['accuracy'],
)

# -- Train -- 
history = model.fit(
    X_train,
    y_train,
    epochs=10, 
    batch_size=512,
    validation_data=[X_val, y_val],
    verbose=2,
)
```

```python
import tensorflow as tf

DROPOUT_AMOUNT = 0.5
REGULARIZATION = 0.01

model2 = keras.Sequential()
# 1) Embedding layer that learns embeddings for each word as we train. 
# The shape of the matrix is len(X[0]) x (desired_emb_dimension) == input_dim x output_dim == (rowsXcols)
model2.add(keras.layers.Embedding(
    input_dim=len(w2i_dict),
    output_dim=20,
     embeddings_initializer=initializers.glorot_normal(seed=None),
#     embeddings_regularizer=regularizers.l2(0.01),
  #   activity_regularizer=regularizers.l1(REGULARIZATION),
))

# 2) Dropout 
model.add(keras.layers.Dropout(0.1))

# 3) Bi-LSTM 
model2.add(keras.layers.Bidirectional(
    layer=keras.layers.LSTM(
        units=10, # Dimensionality of the output space
#         kernel_regularizer=regularizers.l1(0.01),
#         recurrent_regularizer=regularizers.l2(0.01),
        dropout=DROPOUT_AMOUNT,
        recurrent_dropout=DROPOUT_AMOUNT,
#         return_sequences=False,
#         unit_forget_bias=True,
        
    ),
    merge_mode="concat") # Concat the output from fwd & bwd lstm's 
)

# 4) Dropout
model.add(keras.layers.Dropout(DROPOUT_AMOUNT))

# 5) Fully connected layer
model2.add(keras.layers.Dense(
    units=4, 
    activation=tf.nn.relu,
#     kernel_initializer=initializers.glorot_normal,
#     kernel_regularizer=regularizers.l2(0.01),
#     activity_regularizer=regularizers.l1(0.01),
))

# 6) Dropout
model2.add(keras.layers.Dropout(DROPOUT_AMOUNT))

# 7) Fully connected layer
model2.add(keras.layers.Dense(
    units=1, 
    activation="sigmoid",
))


# -- Compile --
model2.compile(
    optimizer="adam",
    loss="binary_crossentropy",
    metrics=['accuracy'],
)

# -- Train -- 
history2 = model2.fit(
    X_train,
    y_train,
    epochs=10, 
    batch_size=256,
    validation_data=[X_val, y_val],
    verbose=2,
)

```
